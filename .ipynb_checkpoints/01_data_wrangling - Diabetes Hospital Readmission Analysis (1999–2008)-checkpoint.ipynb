### Sorting Columns by Missingness and Visualizing Missing Data

In this section, we sort the columns of `df2` based on the **percentage of missing values** in descending order.  
This helps us quickly identify the fields with the highest data gaps.

Next, we visualize the missingness pattern using **Missingno's matrix plot**, which provides a clear picture of the distribution and concentration of missing values across the dataset.

We also create a `missing_data` summary table showing:
- The total number of missing values per column  
- The percentage of missing values  
- Sorted results (descending) to compare with the visual output

Both methods — column sorting and the `missing_data` table — confirm the same ranking of missingness.


```python
missing_data = pd.DataFrame({
    'Missing Count': df2.isnull().sum(),
    'Missing %': (df2.isnull().sum() / len(df2) * 100)
})

missing_data.sort_values('Missing %', ascending=False).head(10)
```

```python
# Sort columns by percentage of missing values (descending)
df2 = df2[df2.isnull().mean().sort_values(ascending=False).index]

# Visualize missingness in our original dataframe 
msno.matrix(df2)
plt.show()
```

### Missing Data Overview

The table above shows the **missing count** and **percentage of missing values** for key features in the dataset.  
This helps us quickly identify columns with severe data gaps that may require removal, imputation, or further review.

- **Extremely high missingness (>80%)** in `weight`, `max_glu_serum`, and `A1Cresult` indicates these features contribute little usable information.
- **Moderate missingness (30–50%)** appears in `medical_specialty` and `payer_code`, suggesting possible targeted imputation or category grouping.
- **Low missingness (<5%)** in variables such as `race`, `diag_3`, `diag_2`, and `diag_1` makes them suitable for straightforward cleaning.

This summary clearly highlights which columns require the most attention during data cleaning and preprocessing.


### Justification for Dropping Highly-Missing Clinical Columns
The columns `weight`, `max_glu_serum`, and `A1Cresult` have extremely high missing 
values (>80%) and even though weight is a risk factor that will be useful in the prediction but due to over 95% of this data is missing and these fields were inconsistently captured in the source EHR and do not offer stable clinical signal for machine learning. Therefore, they are removed here.

```python
# Columns to drop due to high missing values above 80%

cols_to_drop = missing_data[missing_data['Missing %'] > 80].index
cols_to_drop
```

```python
#Columns with only one constant value (or almost constant) add no predictive power

constant_cols = [col for col in df2.columns if df2[col].nunique() <= 1]
constant_cols
```

```python
# Dropping the two variables

all_drop_cols = list(cols_to_drop) + list(constant_cols)
df3 = df2.drop(columns= all_drop_cols)
```

### Handling `payer_code`

`payer_code` reflects a patient’s insurance or funding source, which can influence
access to follow-up care and likelihood of readmission. Although it contains missing
values, dropping the column would remove potentially important socio-economic
information.

Instead, missing entries are replaced with `"Unknown"` so the model can still learn
from this feature without losing data or introducing bias.

```python
#Payercode was removed since it had a high percentage of missing values and it was not considered relevant to the outcome

df3['payer_code'] = df3['payer_code'].fillna('Unknown')
df4 = df3.copy()
```

```python
### Verifying `payer_code` Missing-Value Treatment

df3.payer_code.unique()
```

### For Missing Categorical Values We Filled With "Unknown"
All missing values in `medical_specialty`, `race`, and the diagnosis fields are
treated as `"Unknown"` because these columns are categorical labels, not numeric
measurements. Imputing them using statistical or algorithmic methods would create
clinically incorrect or fabricated values (e.g., guessing diagnoses or race),
which introduces bias and misinformation. 

Missingness in these fields is itself informative—for example, an unassigned
specialty or missing diagnosis often reflects specific encounter characteristics
that influence readmission risk. Using `"Unknown"` preserves this signal without
distorting the data.


```python
# Missing data treatment after cleaning

# High-missing categorical fields
df4['medical_specialty'] = df4['medical_specialty'].fillna('Unknown')
df4['race'] = df4['race'].fillna('Unknown')
```

```python
#Viewing to confirm 'Unknown' as part of the distinct values in race

df4.race.unique()
```

### Removing Terminal/Hospice Discharges
Encounters ending in hospice or death cannot lead to readmission and are removed
to ensure a valid prediction target.


```python
# Excluding hospice patients and dead patients using the ids mapping discharge_disposition_id
 
remove_codes = [11, 13, 14, 19, 20, 21]
# These explicitly indicate the patient that expired OR discharged to a hospice
df5 = df4.loc[~df4['discharge_disposition_id'].isin(remove_codes)]
```

```python
# Verifying 'discharge_disposition_id' column to confirm that remove_codes = [11, 13, 14, 19, 20, 21] are now gone

sorted(df5.discharge_disposition_id.unique().tolist())
```

```python
# keeping all encounters of patient 

df_clean = df5.copy()    
df_clean.head()
```

### Creating the Binary 30-Day Readmission Target

The original `readmitted` column contains three categories ("<30", ">30", "NO"),
but our problem focuses specifically on predicting **30-day readmission risk**—
the clinically meaningful and operationally relevant outcome used in hospitals.

To make the target suitable for machine learning, we convert it into a binary
feature:

- `1` → readmitted within 30 days (`"<30"`)
- `0` → not readmitted within 30 days (`">30"` or `"NO"`)

This binary target aligns with hospital quality metrics, simplifies model
training, and supports accurate prediction of early, avoidable readmissions.


```python
# Create binary 30-day readmission target
df_clean['readmit_30d'] = (df_clean['readmitted'] == '<30').astype(int)

print(df_clean['readmitted'].value_counts(dropna=False))
print(df_clean['readmit_30d'].value_counts())
```

## Age Feature Engineering (Dual Encoding)
We keep both:
- the original age category (10-year bands)
- a numeric midpoint (e.g., `[50-60)` → `55`)

This preserves the raw data while supporting modeling flexibility and interpretability.


```python
df_clean['age_range'] = df_clean['age']

age_mid_map = {
    '[0-10)':5,'[10-20)':15,'[20-30)':25,'[30-40)':35,
    '[40-50)':45,'[50-60)':55,'[60-70)':65,'[70-80)':75,
    '[80-90)':85,'[90-100)':95
}

df_clean['age_mid'] = df_clean['age_range'].map(age_mid_map)
df_clean[['age_range','age_mid']].head()
```

```python
#Diagnosis before Encoding to 3-Digit ICD Extraction
print(df_clean['diag_1'].head())
print(df_clean['diag_2'].head())
print(df_clean['diag_3'].head())
```

### Diagnosis Encoding (3-Digit ICD Extraction)
We extract the first three digits of each diagnosis code. This reduces complexity
while preserving the clinical meaning of ICD-9 categories without depending on
external mappings not included in the dataset.


```python
df_clean['diag_1_3digit'] = df_clean['diag_1'].str[:3]
df_clean['diag_2_3digit'] = df_clean['diag_2'].str[:3]
df_clean['diag_3_3digit'] = df_clean['diag_3'].str[:3]
```

```python
print(df_clean['diag_1_3digit'].head())
print(df_clean['diag_2_3digit'].head())
print(df_clean['diag_3_3digit'].head())
```

```python
# Handling the missing data for Diagnosis fields (raw + 3-digit ICD)

diag_cols = ['diag_1','diag_2','diag_3',
             'diag_1_3digit','diag_2_3digit','diag_3_3digit']

for col in diag_cols:
    df_clean[col] = df_clean[col].fillna('Unknown')
```

### Medication Feature Engineering
Each medication field contains categories such as "No", "Steady", "Up", and "Down".
To capture both treatment presence and dosage adjustments—both clinically relevant
signals for readmission risk—we generate three features per medication:

1. **Binary flag (`_flag`)** — whether the medication was used (1) or not (0).  
2. **Ordinal dosage encoding (`_dose`)** — No=0, Steady=1, Up=2, Down=3.  
3. **Dose change indicator (`_dose_change`)** — identifies if dosage increased or
   decreased (Up/Down = 1), indicating treatment instability.

This combined approach retains clinical detail while improving machine learning
performance and model interpretability.


```python
# Medication Feature Engineering

# List of all diabetes-related medication fields
drug_cols = [
    'metformin','repaglinide','nateglinide','chlorpropamide','glimepiride',
    'acetohexamide','glipizide','glyburide','tolbutamide','pioglitazone',
    'rosiglitazone','acarbose','miglitol','troglitazone','tolazamide',
    'glyburide-metformin','glipizide-metformin','glimepiride-pioglitazone',
    'metformin-rosiglitazone','metformin-pioglitazone','insulin'
]

# Mapping for dosage categories
dose_map = {'No': 0, 'Steady': 1, 'Up': 2, 'Down': 3}

for c in drug_cols:
    if c in df_clean.columns:
        
        # 1. Medication presence (binary flag)
        df_clean[c + '_flag'] = (~df_clean[c].isin(['No']) & df_clean[c].notna()).astype(int)
        
        # 2. Ordinal dosage pattern
        df_clean[c + '_dose'] = df_clean[c].map(dose_map)
        
        # 3. Dose change indicator (Up/Down)
        df_clean[c + '_dose_change'] = df_clean[c].isin(['Up', 'Down']).astype(int)
```

### Outlier Treatment
No outlier removal or capping was applied. Most variables in the dataset are
categorical and therefore cannot have outliers. The numeric variables (e.g.,
`num_lab_procedures`, `num_medications`, `number_inpatient`) reflect real
clinical variation, where higher values indicate greater disease severity rather
than data errors. Removing or capping these values would discard legitimate
high-risk cases and reduce model performance. Therefore, all clinically plausible
numeric values were retained as-is.

# **Data Wrangling Summary**

## **Overview**
We transformed the raw UCI Diabetes 1999–2008 dataset into a clean, structured, and machine-learning-ready dataset for predicting 30-day hospital readmission. The original data contained extensive missing values, coded categorical fields, diagnosis codes, inconsistent medication formats, and no direct readmission target.

## **Key Cleaning Actions**
- Standardized missing values and corrected identifier data types.  
- Removed unusable or high-missingness fields (`weight`, `max_glu_serum`, `A1Cresult`, `examide`, `citoglipton`).  
- Imputed missing categorical values (`race`, `medical_specialty`, diagnosis codes) with `"Unknown"` to preserve meaningful missingness.  
- Removed hospice/terminal discharge dispositions and retained all remaining encounters.  
- Created a clear binary 30-day readmission target from the raw `readmitted` field.

## **Feature Engineering Highlights**
- Added dual age representation (original age bands and numeric midpoint).  
- Extracted 3-digit ICD-9 diagnosis codes to reduce cardinality while preserving clinical meaning.  
- Engineered medication features capturing treatment presence, dosage patterns, and dose-change indicators.

## **Final Outcome**
The result is a clean, reliable, and clinically meaningful dataset that handles missingness appropriately and includes strong engineered features. This dataset is now ready for exploratory data analysis and machine-learning modeling focused on 30-day hospital readmission prediction.
